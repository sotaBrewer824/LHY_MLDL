```python
resnet_path = os.path.join(GOOGLE_DRIVE_PATH, 'resnet18.csv')
vgg_path = os.path.join(GOOGLE_DRIVE_PATH, 'vgg16.csv')
efficientnet_path = os.path.join(GOOGLE_DRIVE_PATH, 'efficientnet.csv')
```

```python
#df1 = pd.read_csv('80585-1.csv', index_col = 0)
#df2 = pd.read_csv('80585-2.csv', index_col = 0)
#df3 = pd.read_csv('predict_deep_transform_semi.csv', index_col = 0)
df1 = pd.read_csv(resnet_path, index_col = 0)
df2 = pd.read_csv(vgg_path, index_col = 0)
df3 = pd.read_csv(efficientnet_path, index_col = 0)
```

```python
df_combine = pd.concat([df1, df2, df3],axis=1, )
df_combine = df_combine.mode(axis=1).dropna(axis=1)
```

```python
df_combine = df_combine.astype('int32')
df_combine.columns = ['Category']
df_combine
```

```python
df_combine.to_csv('Ensemble1.csv',index=True)
```
